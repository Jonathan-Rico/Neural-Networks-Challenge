## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../deep-learning-challenge/Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial columns, 'EIN'.

application_df = application_df.drop(columns=['EIN'])
application_df.head()
```

```python
# Determine the number of unique values in each column.

application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning

application_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

namecount = application_df['NAME'].value_counts()
application_names_to_replace = list(namecount[namecount<5].index)
# Replace in dataframe
for app in application_names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_value_counts = application_df.CLASSIFICATION.value_counts()
classification_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts[classification_value_counts>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_value_counts[classification_value_counts < 800].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_application_df = pd.get_dummies(application_df)
dummies_application_df
```

```python
# Split our preprocessed data into our features and target arrays

y = dummies_application_df['IS_SUCCESSFUL'].values
X = dummies_application_df.drop(['IS_SUCCESSFUL'], axis = 'columns').values
# Split the preprocessed data into a training and testing dataset

X_train,X_test, y_train, y_test = train_test_split(X,y, random_state=1) 
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

neural_net = tf.keras.models.Sequential()

# First hidden layer

neural_net.add(
    tf.keras.layers.Dense(units=100, input_dim = 454, activation = 'relu')
)
# Second hidden layer

neural_net.add(
     tf.keras.layers.Dense(units=30, activation= 'sigmoid')
)
#Third hidden layer

neural_net.add(
    tf.keras.layers.Dense(units=10, activation= 'sigmoid')
)
#Fourth hidden layer

neural_net.add(
    tf.keras.layers.Dense(units=5, activation= 'sigmoid')
)
# Output layer
neural_net.add(
    tf.keras.layers.Dense(units=1,activation='sigmoid')
)
# Check the structure of the model
neural_net.summary()
```

```python
# Compile the model

neural_net.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model_fit = neural_net.fit(X_train_scaled,y_train, epochs =100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = neural_net.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
